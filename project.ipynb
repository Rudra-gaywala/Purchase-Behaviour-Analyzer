## 3. Outlier Handling

```python
num_cols = df1.select_dtypes(include=['int64', 'float64']).columns

for num in num_cols:
    z_scores = (df1[num] - df1[num].mean()) / df1[num].std()
    df4 = df1[(z_scores < 3) & (z_scores > -3)]

print(df4.shape)
```

## IQR Method

```python
for num in num_cols:
    q3 = df1[num].quantile(0.75)
    q1 = df1[num].quantile(0.25)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    df5 = df1[(df1[num] >= lower_bound) & (df1[num] <= upper_bound)]
    print(df5.shape)
```

IQR method is more relavent than the z-score method because it is more accurate than z-score method

## Windsorization

```python
for num in num_cols:
    lower = np.percentile(df1[num], 5)
    upper = np.percentile(df1[num], 95)
windsorize = np.clip(df1[num], lower, upper)
print(windsorize)
```

## 6. Feature Construction

```python
df2['date'] = pd.to_datetime(df2['date'])

df2['year_month'] = df2['date'].dt.to_period('M')

agg = df2.groupby('user_id').agg(
    total_spend=('amount', 'sum'),
    active_months=('year_month', 'nunique')
).reset_index()

agg['avg_monthly_spend'] = agg['total_spend'] / agg['active_months']

agg
```

```python
agg = df2.groupby('user_id').agg(
    total_transactions=('date', 'count'),
    active_months=('year_month', 'nunique')
).reset_index()

agg['purchase_frequency'] = agg['total_transactions'] / agg['active_months']

agg
```

```python
current_date = df2['date'].max()

agg = df2.groupby('user_id').agg(
    last_purchase_date=('date', 'max')
).reset_index()

agg['days_since_last_purchase'] = (current_date - agg['last_purchase_date']).dt.days

agg
```
